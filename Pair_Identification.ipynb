# Background and approach:

Theoretically, two time series are considered as a "pair" if they are mathematically cointegrated i.e. there exists a linear combination of the two series which is "stationary". Therefore, we can set two conditions to find pairs:
1) They show good correlation i.e. a linear model can be fitted between the two series.
2) The residual of the fitting is "stationary" i.e. it has no trend or periodicity

To apply the first condition, linear regression - i.e. baseline supervised machine learning model - can be applied to calculate the correlations between various time series. And, for the second condition, the "augmented Dickeyâ€“Fuller" test (ADF) statistical test can be used to calculate the test p-value. Finally, an unsupervised machine learning model can be used to select pairs with noticeably high correlation and low p-value.

# Forming the dataframe of the price history for the selected symbols

```python
# create empty dataframe
stocks = pd.DataFrame()
# iterate over each symbol
for symbol in symbols:  
    
    # print the symbol which is being downloaded
    print( str(symbols.index(symbol)) + str(':') + symbol, sep=',', end=' , ', flush=True)  
    
    try:
        # download the stock price 
        stock = []
        stock = yf.download(symbol,start=start, end=end, progress=False)
        stock = stock.drop(columns=['Open','High','Low','Adj Close','Volume'])
        stock = stock.rename(columns = {"Close":symbol})
        # append the individual stock prices 
        if len(stocks) == 0:
            stocks = stock
        else:
            stocks = pd.merge(stocks, stock, on="Date", how="outer") #.append(stock,sort=False)
    except Exception:
        None
```

```python
stocks.head()
```

# Cleaning and transforming the stocks price data

```python
stocks.isnull().sum()
```

```python
stocks = stocks.dropna()
```

For our study, we consider the close price of the selected symbols and apply the log transformation. The reason is stocks are "compounding" investments.

```python
stocks_log = np.log(stocks)
stocks_log.head()
```

Calculating the "standard deviations" which we need for calculating correlations:

```python
rss = stocks_log.std()
```

Here, for any pair - without repetition - we fit a linear model between any two set of data, calculate the correlation using the linear coefficient and we also calculate the p-value of the residual. We form another dataframe consisting of the correlation and p-value:

```python
pairs = pd.DataFrame(columns = ['correlation', 'p-value'])
for i1,symbol1 in enumerate(symbols):
    for i2,symbol2 in enumerate(symbols):
        if i2<=i1: continue
        stock1 = stocks_log[[symbol1]]
        stock2 = stocks_log[[symbol2]]
        model = LinearRegression()
        model.fit(stock1,stock2)
        corr =  rss[i1]/rss[i2] * float(model.coef_)
        res = stock2 - model.predict(stock1)
        result = adfuller(res.values)
        p_val = result[1]
        values = {'correlation': corr, 'p-value': p_val}
        row_to_add = pd.Series(values, name=f'{symbol1}_{symbol2}')
        pairs = pairs.append(row_to_add)
        
```

# Unsupervised clustering of the pairs

To find the optimum "epsilon parameter" i.e. the maximum distance between two neighbour object, we need to do an elbow-plot study:

```python
X = pairs.to_numpy()
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
```

```python
plt.plot(distances)
plt.yscale('log')
plt.ylabel('eps parameter')
plt.xlabel('distance')
plt.rcParams['figure.figsize'] = (10,20)
plt.show()
```

The optimum eps related to the turning point can is:

```python
opt_eps = 0.02
```

```python
def plot_clusters(X,y_res, plt_cluster_centers = False):
    X_centroids = []
    Y_centroids = []

    for cluster in set(y_res):
        x = X[y_res == cluster,0]
        y = X[y_res == cluster,1]
        X_centroids.append(np.mean(x))
        Y_centroids.append(np.mean(y))

        plt.scatter(x,
                    y,
                    s=50,
                    marker='s',
                    label=f'cluster {cluster}')

    if plt_cluster_centers:
        plt.scatter(X_centroids,
                    Y_centroids,
                    marker='*',
                    c='red',
                    s=250,
                    label='centroids')
    plt.legend()
    plt.grid()
    plt.rcParams['figure.figsize'] = (10,10)
    plt.ylabel('P-Value')
    plt.xlabel('Correlation')
    plt.show()

```

We use DBSCAN for clustering and plot the labelled data points:

```python
db = DBSCAN(eps= opt_eps, min_samples=3, metric = 'euclidean')
y_db = db.fit_predict(X)
plot_clusters(X,y_db)
```

We can associate the cluster 1 (high correlation and low p-value) with the cointegration definition: 

```python
cluster1 = y_db==1
cointegrated_pairs = pairs[cluster1]
cointegrated_pairs
```
